## Parameter

```python
gap_factor=20 ## num of detection within each second
match_ratio=0.65 ##the distance ratio thershold between best match and second match
good_match_rate=1 ##the rate of matches to be chosen in all
ransacReprojThreshold=6
min_match_count=10
```

## get action sequence

```python
def get_action():
    out_sequence=[]
    ##load test video
    video=os.path.abspath(os.curdir)+'/demo.mp4'
    videoCapture = cv2.VideoCapture(video)
    
    ##detect for gap_factor #gap_dected frames
    fps = videoCapture.get(cv2.CAP_PROP_FPS)
    #print(fps)
    gap_dected=int(fps//gap_factor)
    
    ##load first frame
    success, frame_old = videoCapture.read()
    i,j=0,0
#     _SaveImage(frame_old[80:frame_old.shape[0]-40,:],os.path.abspath(os.curdir)+'/test',j)
    #print('save image:',i)
    
#     predict=deque([],maxlen=gap_factor)
#     for _ in range(gap_factor):
#         predict.append(None)
    
    while success:
        i+=1
    ##finish process if all frames loaded
        success, frame_new = videoCapture.read()
        if success==False:
            print('detection finish')
#           sys.exit() 
            return out_sequence
        if (i % gap_dected == 0):
            j+=1
#             _SaveImage(frame_new[80:frame_new.shape[0]-40,:],os.path.abspath(os.curdir)+'/test',j)
        #    print('save image:',i)
            
    ##predict 
            if _AlignImages(frame_old,frame_new,i) != None:
                predict=_AlignImages(frame_old,frame_new,i)
               
                #print(predict)
                
                res=_Predict(predict)
                out_sequence.append(res)
               # print(i//30,'s',i%30,'frame',res)
            
            
        frame_old=frame_new
    
            
        
        
```

## Helper

```python
def _SaveImage(image,addr,num):
    address = addr +'/'+ str(num)+ '.jpg'
    cv2.imwrite(address,image)

def _AlignImages(img1,img2,filenum):
    # initiate ORB detector
    orb = cv2.ORB_create(5000)
    
    ##check bilateral banner, eliminate stick part
#     up_idx,down_idx=_SplitImage(img1,img2)
#     img1 = img1[up_idx:down_idx,:]
#     img2= img2[up_dx:down_idx,:]
    img1 = img1[80:img1.shape[0]-40,:]
    img2 = img2[80:img2.shape[0]-40,:]
    img2=cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
    img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2GRAY)
    count=0
    
    for i in img2.sum(axis=1):
        if i==0:
            count+=1
    if count/img2.sum(axis=1).shape[0]>0.2:
        return 'open','picture'
    
    ##find the keypoints and descriptors with ORB
    kp1, des1 = orb.detectAndCompute(img1,None)
    kp2, des2 = orb.detectAndCompute(img2,None)
    if des2 is None: 
        return 'click','in'
    
    ##match with Brute-Force
    bf = cv2.BFMatcher(cv2.NORM_HAMMING)
    ##find best 2 matches for each kp
    matches = bf.knnMatch(des1, trainDescriptors = des2, k = 2)
#     print(len(matches))
    ##ratio test
#     debug=[e for e in matches ]
#     with open('debug.txt','a') as file0:
#         print(debug,file=file0)
#         print('\n',file=file0)
    
    if len(matches)==0 or len(matches[0])==1 :
        return 'click','in'
    
    else:
        good = [m for (m,n) in matches if m.distance < match_ratio*n.distance]
    ##sort and choose the best part of matches
#     good = sorted(good, key = lambda x:x.distance)
#     good = good[:int(len(good) * good_match_rate)]
#     img3 = cv2.drawMatches(img1,kp1,img2,kp2,good[int(0.75*len(good)):int(1*len(good))],None,flags=2)
#     _SaveImage(img3,os.path.abspath(os.curdir)+'/compare',filenum)
    
    if len(good)>min_match_count:
        src_pts = np.float32([ kp1[m.queryIdx].pt for m in good[int(0.75*len(good)):int(1*len(good))] ]).reshape(-1,1,2)
        dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good[int(0.75*len(good)):int(1*len(good))]]).reshape(-1,1,2)
        
        src_pts,dst_pts=_Recheck(src_pts,dst_pts,img1.shape)
        if dst_pts.shape[0]<=50:
            return 'click','in'
        res1,res2=_Good_inter_check(src_pts,dst_pts)
        if res1!=0:
            return res1,res2
        ##find (3x3) transform matrix
        M, _ = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,ransacReprojThreshold)
        t_x = M[0][2]
        t_y = M[1][2]
        return (t_x,t_y)
    else:
        return None
        

def _Recheck(src_pts,dst_pts,fig_shape):
    s=[]
    d=[]
    for i in range(src_pts.shape[0]):
        sc=src_pts[i]
        ds=dst_pts[i]
        dx=abs(sc[0][0]-ds[0][0])
        dy=abs(sc[0][1]-ds[0][1])
        if dx/fig_shape[1]<0.3 and dy/fig_shape[0]<0.3:
            s.append(sc)
            d.append(ds)
    s=np.array(s)
    d=np.array(d)
#     print(s.shape)
    return s,d
    
        
def _Good_inter_check(src_pts,dst_pts):
    res=[]
#     res2=[]
    for i in range(src_pts.shape[0]):
        s=src_pts[i]
        d=dst_pts[i]
        dy=abs(s[0][1]-d[0][1])
#         dx=abs(s[0][0]-d[0][0])
        res.append(dy)
#         res2.append(dx)
    res.sort()
#     res2.sort()
    allave=np.mean(res)
    lastave=np.mean(res[9*int(len(res)/10):-1])
#     allave2=np.mean(res2)
#     lastave2=np.mean(res2[9*int(len(res2)/10):-1])
    if (lastave>=5*allave and allave>=2):
        return 'buy good','change'
#     if (lastave2>=5*allave2 and allave2>=2):
#         return 'good picture','change'
    else:
        return 0,0
    
        
        

def _Predict(predict):
    
    res= __Pattern(predict[0],predict[1])
    return res
    
def __Pattern(t_x,t_y):
    if type(t_x)!=str and type(t_y)!=str:
        res=[]
        if abs(t_x)<1 and abs(t_y)<1:
            return 'stop'
        if abs(t_y/t_x)>3:
            if (t_y>3):
                res.append('down')
            if (t_y<-3):
                res.append('up')
        
        if abs(t_x/t_y)>3:
            if (t_x>3):
                res.append('right')
            if(t_x<-3):
                res.append('left')
        if len(res)==0:
            return 'stop'
        else:
            major={'up':0,'down':0,'right':0,'left':0,'stop':0}
            for e in res:
                major[e]+=1
            for k, v in major.items():
                if v == max(major.values()):
                    return k        
        
    else:
#         if t_x=='click' and t_y=='in':
        return t_x+t_y
        
```

## turn into idx sequence

```python
def logic(out_sequence):
    my_dict={'stop':0,'up':1,'down':2,'left':3,'right':4,'clickin':5,'open picture':6,'buy good change':7}
    left_idx_list=[]
    res=[]
    click_symbol=0
    click_add=0
    back_add=0
    for i in range(len(out_sequence)):
        if out_sequence[i]=='stop':
            click_add=0
            back_add=0
            if click_symbol==1:
                left_idx_list=[]
                click_symbol=0
            elif click_symbol==0:
                if len(left_idx_list)!=0:
                    for i in range(len(left_idx_list)):
                        res.append(my_dict['open picture'])
                    left_idx_list=[]
            res.append(my_dict['stop'])
        if out_sequence[i]=='left':
            click_add=0
            back_add=0
            left_idx_list.append(i)
        if out_sequence[i]=='clickin':
            click_symbol=1
            if click_add==0:
                res.append(my_dict['clickin'])
                click_add=1
        if  out_sequence[i]=='right':
            if back_add==0:
                res.append(my_dict['right'])
                back_add=1
        if  out_sequence[i]=='up' or out_sequence[i]=='down' or out_sequence[i]=='buy good change':
            res.append(my_dict[out_sequence[i]])
    return res
            
            
        
                
                
            
```

```python
sequence=logic(out_sequence)
```

## state machine

```python
class shopping(object):
    states=['market1','market2', 'product_page', 'exit1','exit2']
    def __init__(self,name,fps,action):
        self.name=name
        self.fps=fps
        self.report=0
        self.movement={0:'stay',1:'scroll_up',2:'scroll_down',3:'scroll_left',4:'scroll_right',5:'tap_in',6:'view_produ_pic',7:'click_and_buy'}
        self.action=action
        self.list_main=[]
        self.list_product=[]
        self.click_num=0
        self.stay_main_time=0
        self.stay_product_time=0
        
        #initializer
        self.machine=Machine(model=self,states=shopping.states,initial='market1')
        #some transition
            #stay
        self.machine.add_transition(trigger='stay',source='market1',dest='market1')
        self.machine.add_transition(trigger='stay',source='market2',dest='market2')
        self.machine.add_transition(trigger='stay',source='product_page',dest='product_page')
        self.machine.add_transition(trigger='stay',source='exit1',dest='exit1')
        self.machine.add_transition(trigger='stay',source='exit2',dest='exit2')
            #up,down
        self.machine.add_transition(trigger='scroll_up',source='market1',dest='market1')
        self.machine.add_transition(trigger='scroll_up',source='market2',dest='market2')
        self.machine.add_transition(trigger='scroll_up',source='product_page',dest='product_page')
        self.machine.add_transition(trigger='scroll_up',source='exit1',dest='exit1')
        self.machine.add_transition(trigger='scroll_down',source='market1',dest='market1')
        self.machine.add_transition(trigger='scroll_down',source='market2',dest='market2')
        self.machine.add_transition(trigger='scroll_down',source='product_page',dest='product_page')
        self.machine.add_transition(trigger='scroll_down',source='exit1',dest='exit1')
            #left right
        self.machine.add_transition(trigger='scroll_right',source='market1',dest='exit2')
        self.machine.add_transition(trigger='scroll_right',source='market2',dest='market1')
        self.machine.add_transition(trigger='scroll_right',source='product_page',dest='market2')
        self.machine.add_transition(trigger='scroll_right',source='exit1',dest='product_page')
        self.machine.add_transition(trigger='scroll_left',source='market1',dest='market2')
        self.machine.add_transition(trigger='scroll_left',source='market2',dest='market2')
        
            #tap_in
        self.machine.add_transition(trigger='tap_in',source='market1',dest='product_page')
        self.machine.add_transition(trigger='tap_in',source='market2',dest='product_page')
        self.machine.add_transition(trigger='tap_in',source='product_page',dest='exit1')
        self.machine.add_transition(trigger='tap_in',source='exit1',dest='exit2')
            #view_produ_pic
        self.machine.add_transition(trigger='view_produ_pic',source='product_page',dest='product_page')
        self.machine.add_transition(trigger='view_produ_pic',source='exit1',dest='exit1')
        self.machine.add_transition(trigger='view_produ_pic',source='market1',dest='market1')
        self.machine.add_transition(trigger='view_produ_pic',source='market2',dest='market2')
            #click_and_buy
        self.machine.add_transition(trigger='click_and_buy',source='product_page',dest='product_page')
        self.machine.add_transition(trigger='click_and_buy',source='exit1',dest='exit1')

        self.machine.on_enter_market1('enter_market1')
        self.machine.on_enter_market2('enter_market2')
        self.machine.on_enter_product_page('enter_product_page')
        self.machine.on_enter_exit1('enter_exit1')
        self.machine.on_enter_exit2('enter_exit2')
    def enter_market1(self):
        self.stay_main_time+=1
        if self.stay_product_time!=0:
            self.list_product.append(self.stay_product_time)
            self.stay_product_time=0
    def enter_market2(self):
        self.stay_main_time+=1
        if self.stay_product_time!=0:
            self.list_product.append(self.stay_product_time)
            self.stay_product_time=0
    
    def enter_product_page(self):
        self.stay_product_time+=1
        if self.stay_main_time!=0:
            self.list_main.append(self.stay_main_time)
            self.stay_main_time=0
    def enter_exit1(self):
        if self.stay_product_time!=0:
            self.list_product.append(self.stay_product_time)
            self.stay_product_time=0
    def enter_exit2(self):
        pass
#         print('user quit')
    def act(self):
        for i in range(len(self.action)):
            self.trigger(self.movement[self.action[i]])
            if i==(len(self.action)-1):
                if self.stay_product_time!=0:
                    self.list_product.append(self.stay_product_time)
                    self.stay_product_time=0
                if self.stay_main_time!=0:
                    self.list_main.append(self.stay_main_time)
                    self.stay_main_time=0
        for e in self.action:
            if e==5:
                self.click_num+=1
        print('user',self.name,'click',self.click_num,'in this activity.\n')
        print('user',self.name,'stay in main page for',len(self.list_main),'times.\n')
        if len(self.list_main)!=0:
            for i in range(len(self.list_main)):
                print('turn',i+1,' ',self.list_main[i]/self.fps,'s\n')
        print('user',self.name,'stay in product page for',len(self.list_product),'times.\n')
        if len(self.list_product)!=0:
            for i in range(len(self.list_product)):
                print('turn',i+1,' ',self.list_product[i]/self.fps,'s\n')
        if self.state=='exit2':
            if self.report==0:
                print('user quit')
                self.report=1
                
        else:
            print('user stay in page',self.state)
        return
                    
        
        
```
